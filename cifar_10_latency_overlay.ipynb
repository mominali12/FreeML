```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg

# Load the image
image_path = "./latency_plot.png"
img = mpimg.imread(image_path)

# Data normalization
data = np.array([1.6, 14, 15.2, 15.3])
normalized_data = data / 15.3  # Normalize data

data2 = np.array([1.26, 4.24, 8.94, 15.3])
normalized_data2 = data2 / 15.3  # Normalize data
print("Normalized data:", normalized_data)
print("Normalized data2:", normalized_data2)

# X-axis positions for overlayed data
x_positions = np.linspace(0, 1, len(normalized_data))

# Create figure and overlay plot
fig, ax = plt.subplots()

# Display the image
ax.imshow(img, extent=[0, 1, 0, 1], aspect='auto')

# Overlay the graph
ax.plot(x_positions, normalized_data, marker='o', linestyle='-', label="gNet")
ax.plot(x_positions, normalized_data2, marker='o', linestyle='-', label="slimmable")

# Formatting
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xlabel("X-axis")
ax.set_ylabel("Normalized Y-axis")
ax.legend()

# Show plot
plt.show()
```
